```python
from xformers.ops import memory_efficient_attention
from timm.models.vision_transformer import Attention
from functools import partial

def memory_efficient_attention_forward(self, x):
    B, N, C = x.shape
    qkv = self.qkv(x).reshape(B, N, 3, self.num_heads, C // self.num_heads)
    qkv = qkv.permute(2, 0, 3, 1, 4)  # (3, B, heads, N, dim)
    q, k, v = qkv[0], qkv[1], qkv[2]

    out = memory_efficient_attention(q, k, v)
    out = out.transpose(1, 2).reshape(B, N, C)
    return self.proj(out)

def replace_attention_with_xformers(model):
    for module in model.modules():
        if isinstance(module, Attention):
            module.forward = partial(memory_efficient_attention_forward, module)
```

# Download DiT-XL/2 Models

You can choose between a 512x512 model and a 256x256 model. You can swap-out the LDM VAE, too.

```python
image_size = 256 #@param [256, 512]
vae_model = "stabilityai/sd-vae-ft-ema" #@param ["stabilityai/sd-vae-ft-mse", "stabilityai/sd-vae-ft-ema"]
latent_size = int(image_size) // 8
# Load model:
model = DiT_XL_2(input_size=latent_size, in_channels = 4).to(device)
replace_attention_with_xformers(model)
state_dict = find_model(f"DiT-XL-2-{image_size}x{image_size}.pt")
model.load_state_dict(state_dict)
model.eval() # important!
vae = AutoencoderKL.from_pretrained(vae_model).to(device)

import inspect
print(inspect.getsource(model.forward_with_cfg))
```

# 2. Sample from Pre-trained DiT Models

You can customize several sampling options. For the full list of ImageNet classes, [check out this](https://gist.github.com/yrevar/942d3a0ac09ec9e5eb3a).

```python
import time

# Set user inputs:
seed = 0 #@param {type:"number"}
torch.manual_seed(seed)
num_sampling_steps = 250 #@param {type:"slider", min:0, max:1000, step:1}
cfg_scale = 15 # @param {"type":"slider","min":0,"max":15,"step":0.1}
class_labels = 207, 281, 404, 504 #@param {type:"raw"}
samples_per_row = 4 #@param {type:"number"}

# Create diffusion object:
diffusion = create_diffusion(str(num_sampling_steps))

# Create sampling noise:
n = len(class_labels)
z = torch.randn(n, 4, latent_size, latent_size, device=device)
y = torch.tensor(class_labels, device=device)

# Setup classifier-free guidance:
z = torch.cat([z, z], 0)
y_null = torch.tensor([1000] * n, device=device)
y = torch.cat([y, y_null], 0)
model_kwargs = dict(y=y, cfg_scale=cfg_scale)

# Sample images:
start = time.time()
samples = diffusion.p_sample_loop(
    model.forward_with_cfg, z.shape, z, clip_denoised=False,
    model_kwargs=model_kwargs, progress=True, device=device
)
end = time.time()
samples, _ = samples.chunk(2, dim=0)  # Remove null class samples
samples = vae.decode(samples / 0.18215).sample

# Save and display images:
save_image(samples, "sample.png", nrow=int(samples_per_row),
           normalize=True, value_range=(-1, 1))
decoded_samples = samples
samples = Image.open("sample.png")
display(samples)
print(f"⏱️ Time taken to generate 50 images: {end - start:.2f} seconds")
print(f"🧠 Max memory used: {torch.cuda.max_memory_allocated() / 1e9:.2f} GB")
```
